```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv("50_Startups.csv")
data
```

```python
pd.get_dummies(data.State)
```

```python
newdata=pd.get_dummies(data)
newdata
```

```python
newdata.isnull().sum()
```

```python
x=newdata.drop(['Profit'], axis=1)
y=newdata['Profit']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

op= LinearRegression()
op.fit(x_train,y_train)
```

```python
y_pred= op.predict(x_test)

from sklearn import metrics
score_1= metrics.r2_score(y_test,y_pred)
score_1
```

```python
plt.plot(y_test,y_pred)
```
